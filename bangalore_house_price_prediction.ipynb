```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/Bengaluru_House_Data.csv")
data
```

```python
data.drop(columns=['area_type','availability','society'],inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace =True)
```

```python
data
```

```python
data['location'] = data['location'].apply(lambda x: x.strip())
```

```python
data.location.value_counts()
```

```python
location_stats = data.groupby('location')['location'].agg("count").sort_values(ascending=False)
```

```python
location_stats
```

```python
location_less_than_10_enteries = location_stats[location_stats <= 10]
```

```python
location_less_than_10_enteries
```

```python
data['location'] = data['location'].apply(lambda x: 'other' if x in location_less_than_10_enteries else x)
```

```python
data.location.value_counts()
```

```python
data
```

```python
data['size'].unique()
```

```python
data['bedrooms'] = data['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
data
```

```python
data['total_sqft'].unique()
```

```python
def clean(sqft):
  tokens = sqft.split('-')
  if len(tokens) == 2:
    return(float(tokens[0])+float(tokens[1]))/2
  else:
      try:
        return float(sqft)
      except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean)
```

```python
data['total_sqft'].unique()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['sqft_per_bed'] = data['total_sqft']/data['bedrooms']
```

```python
data
```

```python
data.sqft_per_bed.describe()
```

```python
data2 = data[data['sqft_per_bed'] >= 300]
```

```python
data2
```

```python
data2['price_per_sqft'] = data2['price'] * 100000 / data2['total_sqft']
```

```python
data2
```

```python
data2.price_per_sqft.describe()
```

```python
data3 = data2[data2['price_per_sqft'] >= 2000]
```

```python
data3
```

```python
data3.drop(columns=['size','sqft_per_bed','price_per_sqft'],axis=1, inplace=True)
```

```python
data3
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['location']), remainder = 'passthrough')
```

```python
col_trans
```

```python
lr = LinearRegression()
```

```python
scaler = StandardScaler()
```

```python
model = make_pipeline(col_trans, scaler, lr)
```

```python
def remove_outliers_per_location(df):
    cleaned = []
    for loc, group in df.groupby('location'):
        q1, q3 = group['price_per_sqft'].quantile([0.25, 0.75])
        iqr = q3 - q1
        filtered = group[(group['price_per_sqft'] >= q1 - 1.5*iqr) &
                         (group['price_per_sqft'] <= q3 + 1.5*iqr)]
        cleaned.append(filtered)
    return pd.concat(cleaned)
```
